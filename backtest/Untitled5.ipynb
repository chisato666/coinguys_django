```python
#Here's an example of a backtest in Python that checks if the price of Bitcoin increases by 100 USDT in 30 seconds, then sells it if the price increases by 200 USDT:

import time
import ccxt

# Create Binance exchange instance
exchange = ccxt.binance()

# Define trading parameters
symbol = 'BTC/USDT'
buy_threshold = 100
sell_threshold = 200
candle_interval = '1m'
quantity = 0.001

# Get historical price data
ohlcv = exchange.fetch_ohlcv(symbol, candle_interval)

# Loop through historical price data
for i in range(1, len(ohlcv)):
    # Get current and previous candlestick prices
    current_price = ohlcv[i][4]
    previous_price = ohlcv[i - 1][4]
    print(current_price)
    # Check if price has increased by buy threshold
    if current_price - previous_price >= buy_threshold:
        # Buy Bitcoin
        #order = exchange.create_order(symbol, 'market', 'buy', quantity)
        print(f'Bought {quantity} Bitcoin at market price of {current_price}')

        # Wait for price to increase by sell threshold
        while True:
            time.sleep(1)
            current_price = exchange.fetch_ticker(symbol)['last']
            if current_price - previous_price >= sell_threshold:
                # Sell Bitcoin
                #order = exchange.create_order(symbol, 'market', 'sell', quantity)
                print(f'Sold {quantity} Bitcoin at market price of {current_price}')
                break
```
