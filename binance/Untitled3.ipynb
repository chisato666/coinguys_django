````python
 To accomplish this task, you can use the Binance API to get the necessary data. Here's an example Python script that checks for the conditions you mentioned every minute:

```python
import time
from binance.client import Client
from ta.momentum import RSIIndicator
from ta.trend import MACD

# Replace with your Binance API keys
api_key = "your_api_key"
api_secret = "your_api_secret"

# Initialize the Binance client
client = Client(api_key, api_secret)

# Define the symbol to check
symbol = "BTCUSDT"

# Define the time period to check
time_period = "1h"

# Define the RSI and MACD thresholds
rsi_threshold = 70
macd_signal_threshold = 0

# Define the percentage increase threshold
percentage_increase_threshold = 0.01

# Define the initial highest price
highest_price = 0

while True:
    # Get the last candle for the 1-hour time period
    klines = client.get_historical_klines(symbol, time_period, "1 hour ago UTC")
    last_candle = klines[-2]

    # Extract the relevant data
    close_price = float(last_candle[4])
    high_price = float(last_candle[2])
    low_price = float(last_candle[3])
    volume = float(last_candle[5])

    # Check if the current price is the highest in the last 24 hours
    if high_price > highest_price:
        highest_price = high_price

    # Check if the current price increased by the percentage threshold within 1 minute
    current_price_increase = (close_price - low_price) / low_price
    if current_price_increase > percentage_increase_threshold:
    
        # Calculate the RSI and MACD indicators
        close_prices = [float(kline[4]) for kline in klines[:-1]]
        rsi_indicator = RSIIndicator(close_prices)
        macd_indicator = MACD(close_prices)

        # Check if the RSI is above the threshold and the MACD signal is positive
        rsi_value = rsi_indicator.rsi()[-1]
        macd_signal = macd_indicator.macd_signal()[-1]
        if rsi_value > rsi_threshold and macd_signal > macd_signal_threshold:
            print("Sell signal detected!")
    
    # Wait for 1 minute before checking again
    time.sleep(60)
```

This script initializes the Binance API client with your API keys, then defines the symbol and time period to check. It also sets the RSI and MACD thresholds, as well as the percentage increase threshold.

The script then enters an infinite loop where it gets the last candle for the 1-hour time period from the Binance API and extracts the relevant data. It checks if the current price is the highest in the last 24 hours and if it increased by the percentage threshold within 1 minute.

If those conditions are met, the script calculates the RSI and MACD indicators using the `ta` library and checks if the RSI is above the threshold and the MACD signal is positive. If both conditions are met, it prints a "Sell signal detected!" message.

The script then waits for 1 minute before checking again.

````
